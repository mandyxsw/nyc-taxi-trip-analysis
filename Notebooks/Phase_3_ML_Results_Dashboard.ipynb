# Machine Learning Results Dashboard

This notebook provides a lightweight interactive dashboard for the final project’s machine learning results.

The dashboard summarizes three parts of the project:
1. baseline supervised modeling,
2. unsupervised clustering and feature engineering,
3. advanced supervised learning.

This notebook is included as an optional bonus component in order to present the machine learning results in a more compact and interactive format.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, clear_output

# -----------------------------
# Style
# -----------------------------
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# -----------------------------
# Widgets
# -----------------------------
section_selector = widgets.Dropdown(
    options=[
        ("Overview", "overview"),
        ("Baseline Model", "baseline"),
        ("Unsupervised Learning", "unsupervised"),
        ("Advanced Supervised", "advanced_supervised"),
        ("Key Takeaways", "takeaways")
    ],
    value="overview",
    description="Section:",
    style={"description_width": "initial"},
    layout=widgets.Layout(width="340px")
)

output = widgets.Output()

# -----------------------------
# Dashboard renderer
# -----------------------------
def render_dashboard(section):
    with output:
        clear_output(wait=True)

        if section == "overview":
            print("Machine Learning Results Dashboard")
            print("=" * 45)
            print("Dataset: NYC TLC HVFHV trip data")
            print("Target variable: driver_pay")
            print()
            print("This dashboard provides a lightweight interactive summary")
            print("of the project's machine learning results.")
            print()
            print("Included sections:")
            print("- Baseline supervised modeling")
            print("- Unsupervised clustering and feature engineering")
            print("- Advanced supervised learning")
            print()
            print("Main supervised question:")
            print("Can trip distance provide a useful baseline prediction of driver pay?")
            print()
            print("Main unsupervised question:")
            print("Can trip patterns be grouped into meaningful behavioral clusters?")
            print()
            print("This notebook is included as a lightweight bonus interactive dashboard.")

        elif section == "baseline":
            print("Baseline Supervised Modeling")
            print("=" * 45)

            baseline_results = pd.DataFrame({
                "Split": ["Train", "Validation", "Test"],
                "MAE": [5.1231, 5.1208, 5.1232],
                "MSE": [63.8439, 62.1817, 62.3973],
                "RMSE": [7.9902, 7.8855, 7.8992],
                "R2": [0.8010, 0.8051, 0.8052]
            })

            display(baseline_results)

            print("Interpretation:")
            print("The baseline linear regression model performs consistently across")
            print("training, validation, and test sets, with test R² around 0.81.")
            print("This makes it a useful interpretable benchmark for later advanced models.")
            print()

            fig, ax = plt.subplots(figsize=(8, 5))
            ax.bar(
                baseline_results["Split"],
                baseline_results["R2"],
                color=["steelblue", "darkorange", "seagreen"]
            )
            ax.set_title("Baseline Model R² by Split")
            ax.set_ylabel("R²")
            ax.set_ylim(0, 1)
            plt.show()

        elif section == "unsupervised":
            print("Unsupervised Learning and Feature Engineering")
            print("=" * 45)

            cluster_eval = pd.DataFrame({
                "k": [2, 3, 4, 5, 6, 7, 8],
                "silhouette_score": [0.41, 0.36, 0.31, 0.28, 0.25, 0.23, 0.21],
                "calinski_harabasz_score": [42000, 39000, 36000, 34500, 33000, 32000, 31000],
                "davies_bouldin_score": [0.82, 0.91, 1.03, 1.10, 1.16, 1.20, 1.24]
            })

            print("Cluster Evaluation Metrics")
            display(cluster_eval)

            print("The final clustering solution uses k = 2, selected using")
            print("multiple evaluation metrics and later validated on the full cleaned dataset.")
            print()

            cluster_profile = pd.DataFrame({
                "cluster": ["Cluster 0", "Cluster 1"],
                "trip_miles_mean": [2.8, 8.9],
                "trip_time_mean": [820, 1850],
                "driver_pay_mean": [13.5, 33.2],
                "speed_mph_mean": [12.3, 18.5]
            })

            print("Cluster Profile Summary")
            display(cluster_profile)

            fig, ax = plt.subplots(figsize=(8, 5))
            profile_plot = cluster_profile.set_index("cluster")[[
                "trip_miles_mean",
                "trip_time_mean",
                "driver_pay_mean"
            ]]
            profile_plot.plot(kind="bar", ax=ax)
            ax.set_title("Cluster Profile Comparison")
            ax.set_ylabel("Mean Value")
            plt.xticks(rotation=0)
            plt.show()

            print("Interpretation:")
            print("The clustering workflow identifies two broad trip patterns:")
            print("one shorter/lower-paying pattern and one longer/higher-paying pattern.")
            print("Full-dataset validation suggests that this structure remains meaningful at scale.")

        elif section == "advanced_supervised":
            print("Advanced Supervised Learning")
            print("=" * 45)

            validation_results = pd.DataFrame({
                "Model": ["Random Forest", "XGBoost"],
                "MAE": [1.7103, 1.7825],
                "MSE": [14.9082, 17.5312],
                "RMSE": [3.8611, 4.1870],
                "R2": [0.9533, 0.9451]
            })

            print("Validation Set Comparison")
            display(validation_results)

            final_test_results = pd.DataFrame({
                "Best Model": ["Random Forest"],
                "MAE": [1.5952],
                "MSE": [10.7986],
                "RMSE": [3.2861],
                "R2": [0.9662]
            })

            print("Final Test Set Performance")
            display(final_test_results)

            fig, ax = plt.subplots(figsize=(8, 5))
            ax.bar(validation_results["Model"], validation_results["RMSE"], color=["steelblue", "darkorange"])
            ax.set_title("Advanced Supervised Models: Validation RMSE")
            ax.set_ylabel("RMSE")
            plt.show()

            print("Interpretation:")
            print("The advanced supervised models substantially outperform the baseline")
            print("linear regression benchmark. Random Forest achieves the strongest")
            print("validation performance and also produces strong final test results,")
            print("showing the value of richer features and nonlinear modeling.")

        elif section == "takeaways":
            print("Key Takeaways")
            print("=" * 45)
            print("1. Trip distance is a strong predictor of driver pay, and the")
            print("   baseline linear regression model provides a useful interpretable benchmark.")
            print()
            print("2. Unsupervised learning identifies two broad trip patterns,")
            print("   separating shorter/lower-paying trips from longer/higher-paying trips.")
            print()
            print("3. Advanced supervised models substantially improve predictive")
            print("   performance relative to the simple baseline.")
            print()
            print("4. Together, these results show that the project benefits from")
            print("   combining EDA, baseline modeling, unsupervised learning,")
            print("   and advanced supervised methods in a single pipeline.")

# -----------------------------
# Callback
# -----------------------------
def on_section_change(change):
    if change["name"] == "value":
        render_dashboard(change["new"])

section_selector.observe(on_section_change, names="value")

# -----------------------------
# Display dashboard
# -----------------------------
display(section_selector, output)
render_dashboard(section_selector.value)
```
